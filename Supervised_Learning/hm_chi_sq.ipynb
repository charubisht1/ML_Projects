```python
hmdata = pd.read_csv('C:/Users/Study/OneDrive/Desktop/DU/Business_Intelligence/Home_assignment_BI/dataset_hm.csv')
hmdata = hmdata.sample(frac=0.2, random_state=100)
hmdata.describe(include=object)
hmdata.shape
hmdata.info() 
```

Results generated by chi-square are more reliable if values in the table are greater than 5. Based on the output of the first iteration we remove product_group_name, graphical_appearance_name and perceived_colour_value_name which donâ€™t satisfy this condition. We name is dataframe hmdata_chi. 
To avoid the repetition of results we exclude the first iteration of chi-square test from the notebook. 

```python
hmdata_chi = hmdata[ (hmdata["product_group_name"] != "Nightwear") & (hmdata["product_group_name"] != "Swimwear") & (hmdata["product_group_name"] != "Shoes")]  
hmdata_chi = hmdata_chi[(hmdata["graphical_appearance_name"] != "Argyle") & (hmdata["graphical_appearance_name"] != "Contrast") & (hmdata["graphical_appearance_name"] != "Mesh")& (hmdata["graphical_appearance_name"] != "Metallic") & (hmdata["graphical_appearance_name"] != "Neps") & (hmdata["graphical_appearance_name"] != "Other pattern") & (hmdata["graphical_appearance_name"] != "Contrast") & (hmdata["graphical_appearance_name"] != "Slub") & (hmdata["graphical_appearance_name"] != "Transparent") & (hmdata["graphical_appearance_name"] != "Unknown")]  
hmdata_chi = hmdata_chi[(hmdata["perceived_colour_value_name"] != "Undefined")]
hmdata_chi.shape
hmdata_product = hmdata_chi["product_group_name"] 
hmdata_pattern = hmdata_chi["graphical_appearance_name"]
hmdata_color = hmdata_chi["perceived_colour_value_name"]
```

Now we run chi-square test

To check association between hmdata_product, hmdata_pattern

```python
crosstab_pattern, res_pattern, exp_pattern = researchpy.crosstab(hmdata_product, hmdata_pattern, test='chi-square', expected_freqs=True)
# cross table contained observed values
crosstab_pattern
```

```python
#chi-square estimate and p-value
res_pattern
```

```python
#expected values
exp_pattern
```

To check association between hmdata_product, hmdata_color

```python
crosstab_color, res_color, exp_color = researchpy.crosstab(hmdata_product, hmdata_color, test='chi-square', expected_freqs=True)

# cross table contained observed values
crosstab_color
```

```python
#expected values
exp_color
```

```python
#chi-square estimate and p-value
res_color
```

The result shows us that product pattern and product colour have a chi-square estimate of 2909.8 and 485.8 respectively which is much greater than their respective critical values. Hence, we conclude that there is definite association between dependent and independent variables. This perception is further strengthened by the significant p-values generated by the model.
